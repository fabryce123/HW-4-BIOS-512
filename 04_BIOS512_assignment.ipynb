```r
library(readr)
library(dplyr)
library(ggplot2)
library(forcats)
wind_turbine<-read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv",  col_types = cols())
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
ggplot(wind_turbine) +geom_bar(aes(fct_infreq(province_territory)))+coord_flip()
```

Make a bar **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
newdata<-wind_turbine %>% mutate(manufacturer = fct_lump(manufacturer, n = 5)) %>%count(manufacturer)
newdata %>%mutate(manufacturer = fct_reorder(manufacturer, n)) %>% ggplot(aes(x = manufacturer, y = n)) + geom_col() +coord_flip()
```
